```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:/Users/mylaptop/Desktop/weatherHistory.csv")
```

```python
data.rename(columns = {"Apparent Temperature (C)": "Apparent_Temperature"})
data.columns = ['Formatted Date','Summary','Precip Type','Temperature (C)','Apparent_Temperature',
                'Humidity','Wind Speed (km/h)','Wind Bearing (degrees)','Visibility (km)',
                'Pressure (millibars)','Daily Summary'] 
```

```python
data.head()
```

```python
df = data.drop(['Summary','Precip Type','Temperature (C)','Wind Speed (km/h)','Wind Bearing (degrees)','Visibility (km)',
                'Pressure (millibars)','Daily Summary'],axis=1)
```

```python
df
```

```python
df.head()
```

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'])
```

```python
df1 = df.set_index('Formatted Date')
```

```python
monthly_resampled_data = df1.resample('M').mean()
monthly_resampled_data
```

```python

monthly_resampled_data.plot.line(title="Difference between Apparent_Temperature and Humidity")
```

```python
april_resampled_data = monthly_resampled_data[monthly_resampled_data.index.month==4]
april_resampled_data.plot.line(title="Comparision of Apparent_tempareture and Humidity[2006-2016]")
```
